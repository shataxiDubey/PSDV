1. List of first 20 natural numbers

```python
arr = []
for i in range(1, 21):
    arr.append(i)

print(arr)

# sum of even numbers
sum = 0
for element in arr:
    if element%2 == 0:
        sum += element
print(f'Sum of even numbers {sum}')

# product of odd numbers
prod = 1
for element in arr:
    if element%2 != 0:
        prod *= element

print(f'Product of odd numbers {prod}')

# mean, median and standard deviation of the list
sum = 0
for element in arr:
    sum += element
mean = sum / 20
print(f'Mean of 20 numbers {mean}')

median = (arr[10]+arr[11])/2
print(f'Median of 20 numbers {median}')

std = 0
for element in arr:
    std += (element - mean)**2

std = std / 20
print(f'Standard deviation of 20 numbers {std}')
```

3. A list of the first n Fibonacci numbers.

```python
def fib(n, fib_nums = []):
    if n == 0 or n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

```python
n = 10
fib_list = []
for i in range(1, n+1):
    fib_list.append(fib(i))

print(f'Fibonacci list {fib_list}')
```

Plot the fibonacci sequence using the line graph

```python
fig, ax = plt.subplots()
ax.plot(range(1,11), fib_list)
```

4. List Manipulation

```python
farr = [3,5,7,9,11,13]
print(f'Original array {farr}')
farr = farr[:3]+[6]+farr[3:]
print(f'Insert 6 at 3rd index {farr}')
farr.remove(7)
print(f'Remove 7 from farr {farr}')
farr.reverse
print(f'Reverse the list {farr}')
farr.sort()
sum = 0
for x in farr:
    sum += x
n_len = len(farr)
mean = sum / n_len
if n_len % 2 == 0:
    median = (farr[n_len//2]+farr[n_len//2+1])//2
else:
    median = farr[n_len//2]
print(f'Mean {mean}, Median {median}')
```

7. Dictionary operations

```python
student_names = ['Neha','Sneha','Ram', 'Shyam', 'Vinita']
marks = [93,91,90,92,95]
student_dic = {}
for student_name, marks in zip(student_names, marks):
    student_dic[student_name] = marks

print(f'Dictionary of student data {student_dic}')
```

```python
total_marks = 0
for key, val in student_dic.items():
    total_marks += val

avg_marks = total_marks / len(student_dic)
print(f'Average marks: {avg_marks}')

marks_above_avg = []
for name, marks in student_dic.items():
    if marks > avg_marks:
        marks_above_avg.append(name)

print(f'Names of student scoring above average {marks_above_avg}')
```

9. Merging dictionaries

```python
dict1 = {'a': 10, 'b': 20}
print(f'Dictionary 1 {dict1}')
dict2 = {'b': 5, 'c': 15}
print(f'Dictionary 2 {dict2}')

for key, value in dict2.items():
    if key in dict1.keys():
        dict1[key] += value
    else:
        dict1[key] = value

print(f'Merged dictionaries {dict1}')
```

11. Remove duplicate elements

```python
num_arr = [1,2,3,4,2,3,1,4,5]
print(f'Original array {num_arr}')
n = len(num_arr)
tmp_num_arr = []
for element in num_arr:
    if element not in tmp_num_arr:
        tmp_num_arr.append(element)
num_arr = tmp_num_arr
print(f'Array after removing duplicate elements {num_arr}')

n = len(num_arr)
mean = sum(num_arr) / n
if n % 2 == 0:
    median = num_arr[n//2] + num_arr[n//2+1]
else:
    median = num_arr[n//2]

print(f'Median of the unique elements {median},\n Mean of the unique elements {mean}')
```

12. String operations

```python
def find_longest_string(string_list):
    max_index = 0
    for i, element in enumerate(string_list):
        if len(element) > len(string_list[max_index]):
            max_index = i
    
    return string_list[max_index]   
```

```python
string_list = ['abcdefgh', 'ancjdhffjhgfdkjgdfjg', 'yertuieigshgdvjzxncvhgiodjbvhjdsgdafxdhjcsdhfgig', 'ahfdgshf']
max_length_string = find_longest_string(string_list)
print(f'Max length string: {max_length_string}')
```

```python
count = 0
for element in string_list:
    if element[0] in ['a','e','i','o','u']:
        count += 1

print(f'Number of strings starting with vowel {count}')
```

```python
lengths = []
for element in string_list:
    lengths.append(len(element))

print(lengths)
fig, ax = plt.subplots()
ax.hist(lengths, range(0,50))
```

15. Anagram finder

```python
def anagram_finding(word, words):
    anagram = []
    word = word.lower()
    word = [w for w in word]
    word.sort()
    word = ''.join(word)
    new_words = []
    for w in words:
        w = w.lower()
        w = [w for w in w]
        w.sort()
        w = ''.join(w)
        new_words.append(w)

    for i, w in enumerate(new_words):
        if word == w:
            anagram.append(words[i])
    
    return anagram
word = "listen"
words = ["enlist", "google", "inlets", "banana"]
anagram = anagram_finding(word, words)
print(f'Anagram {anagram}')
```

16. Set operations

```python
num_list1 = [1,2,3,4]
num_list2 = [2,3,5,6,1,7]
print(f'Original list {num_list1}, {num_list2}')
num_list1 = set(num_list1)
num_list2 = set(num_list2)
union_list = num_list1.union(num_list2)
intersect_list = num_list1.intersection(num_list2)
sym_diff_list = num_list1.symmetric_difference(num_list2)
print(f'Union {union_list}')
print(f'Intersection {intersect_list}')
print(f'Symmetric difference {sym_diff_list}')
```

18. Word count and frequency

```python
# Read a file and count the number of words in it.
fp = open('words.txt', 'r')
lines = fp.readlines()
all_words = []
for line in lines:
    all_words += line.split(' ')
all_words = [word[:-1] if word[-1] == '\n' else word for word in all_words ]
print(f'No of words in the file {len(all_words)}')
c = Counter(all_words)
freq = c.most_common(1)
print(f'Most frequent word: {freq[0][0]}')

# word frequencies visualisation
fig, ax = plt.subplots()
ax.bar(list(c.values()), list(c.keys()))
```

21. Basic Statistics

```python
import random
num_list = random.choices(range(1,51), k = 100)
num_list.sort()
mean = sum(num_list)/len(num_list)
mode = Counter(num_list)
mode = mode.most_common(1)
mode = mode[0]
n = len(num_list)
if n % 2 == 0:
    median = (num_list[n//2] + num_list[n//2 + 1])/2
else:
    median = num_list[n//2]

diff = 0
for element in num_list:
    diff += (element - mean)**2
variance = diff / len(num_list)
std = variance**0.5
print(f'Mean {mean}, Mode {mode}, Median {median}, Standard deviation {std}')

fig, ax = plt.subplots()
ax.boxplot(num_list)
```

22. Normalization

```python
def norm(num_list):
    tot = sum(num_list)
    normalize = [num/tot for num in num_list]
    return normalize

num_list = [1,2,354,5,653,2245,778,2]
norm_list = norm(num_list)
print(f'Normalized list {norm_list}')


fig, ax = plt.subplots()
ax.plot(norm_list, label = 'Normalized')
ax.plot(num_list, label = 'Unnormalized')
plt.legend()
```
